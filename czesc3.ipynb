```python
#  CZĘŚĆ 3
#  1. Wczytaj dane z plików i przypisz je do zmiennych o wybranych przez siebie nazwach.
#  2. Narysuj dwa wykresy liniowe z wypełnieniem, znajdujące się na jednym rysunku. Wypełnienie
#   powinno być zlokalizować pomiędzy danym wykresem, a odpowiadającą osią poziomą. Niech
#   wykres stali miękkiej narysowany będzie kolorem czerwonym, z przeźroczystym wypełnieniem tego
#   samego koloru, o parametrze przenikania 0.1. Niech wykres stali twardej narysowane będzie kolorem
#   niebieskim, z przeźroczystym wypełnieniem tego samego koloru, o parametrze przenikania 0.1. Dla
#   obu wykresów zakres osi x powinien wynosić od 0 do 3, a jej opis to Przemieszczenie, mm. Zakres osi y 
#   dla obu wykresów powinien wynosić od 0 do 50000, a jej opis to Siła, N. Oba wykresy
#   powinny mieć linie siatki.
#  3. Oblicz zakreślone pole pod wykresem dla obu plików.

# Importowanie potrzebnych bibliotek: numpy - wczytywanie danych,  matplotlib.pyplot - tworzenie wykresów
import numpy as np
import matplotlib.pyplot as plt
```

```python
#  1.Import danych z plików z pomocą biblioteki numpy do zmiennej "dane"

# Pobieranie danych z plików
dane_miekka = np.loadtxt("dane/czesc3/stal_miekka.TXT",delimiter=',')
dane_twarda = np.loadtxt("dane/czesc3/stal_twarda.TXT",delimiter=',')

liczba_wierszy1 = dane_miekka.shape[0]
liczba_wierszy2 = dane_twarda.shape[0]

# Tworzenie list
list_miekka_x = []
list_miekka_y = []

list_twarda_x = []
list_twarda_y = []

# Przypisywanie danych do list
for i in range(0, liczba_wierszy1):
    list_miekka_x.append(dane_miekka[i][0])
    list_miekka_y.append(dane_miekka[i][1])

for i in range(0, liczba_wierszy2):
    list_twarda_x.append(dane_twarda[i][0])
    list_twarda_y.append(dane_twarda[i][1])    
```

```python
# 2. Rysowanie wykresów
fig, ax = plt.subplots(2)

# Wykres na górze
ax[0].set_xlim([0,3])
ax[0].set_ylim([0,50000])
ax[0].grid()
ax[0].plot(list_miekka_x,list_miekka_y,c='r')
ax[0].fill_between(list_miekka_x,list_miekka_y,alpha=0.1,color='r')
ax[0].set_xlabel('Przemieszczenie, mm')
ax[0].set_ylabel('Siła, N')
ax[0].set_title('Stal miękka')

# Wykres na dole
ax[1].set_xlim([0,3])
ax[1].set_ylim([0,50000])
ax[1].grid()
ax[1].plot(list_twarda_x,list_twarda_y,c='b')
ax[1].fill_between(list_twarda_x,list_twarda_y,alpha=0.1,color='b')
ax[1].set_xlabel('Przemieszczenie, mm')
ax[1].set_ylabel('Siła, N')
ax[1].set_title('Stal twarda')
plt.tight_layout()
```

```python
#  3. Obliczanie zakreślonego pola pod wykresem
Pole_miekka = np.trapz(list_miekka_y,list_miekka_x)/1000
Pole_twarda = np.trapz(list_twarda_y,list_twarda_x)/1000
# Komunikat
print('Pole pod wykresem stali miękkkej wynosi:',Pole_miekka,",natomiast pod wykresem stali twardej wynosi:",Pole_twarda)
```
