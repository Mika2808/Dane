```python
#  CZĘŚĆ 1 
#  Dla podanego zbioru punktów, które znajdują się w folderze czesc1, w pliku punkty.TXT wykonaj
#  następujące zadania:
#   1. Wczytaj dane z pliku i przypisz je do zmiennej o wybranej przez siebie nazwie.
#   2. Narysuj wykres punktowy, w formie czerwonych kropek dla co 10 punktu z wczytanego zbioru.
#   3. Przeprowadź aproksymację liniową dla wczytanych danych, a następnie wyświetl współczynnik
#    kierunkowy funkcji aproksymującej.
#   4. Narysuj dwa, nałożone na siebie wykresy. Pierwszy z nich to wykres punktowy dla danego zbioru
#    punktów, a drugi to wykres liniowy dla funkcji aproksymującej. Wykres powinien zawierać opisy
#    osi x oraz y, a także legendę.

# Importowanie potrzebnych bibliotek: numpy - wczytywanie danych,  matplotlib.pyplot - tworzenie wykresów
import numpy as np
import matplotlib.pyplot as plt

#  1.Import danych z plików z pomocą biblioteki numpy do zmiennej "dane"
dane = np.loadtxt("dane/czesc1/punkty.TXT",delimiter=',')
```

```python
# 2.Wykres punktowy, co 10 punkt, za pomocą list_x i list_y i biblioteki matplotlib.pyplot

# Pobieranie ilości wierszy
liczba_wierszy = dane.shape[0] 

# Tworzenie list zawierające współrzędne x i y dla każdego co 10 punktu
list_x = []
list_y = []

# Pobieranie danych co 10 punktu z orginalnej macierzy "dane"
for i in range(0, liczba_wierszy, 10):
    list_x.append(dane[i][0])
    list_y.append(dane[i][1])

# Tworzenie wykresu punktowego z czerwonymi znacznikami
plt.scatter(list_x,list_y,c='red')
```

```python
# 3. Aproksymację liniowa i wyświetlanie współczynnika kierunkowego funkcji aproksymującej
a = np.polyfit(dane[:,0],dane[:,1],1)

print("Współczynik kierunkowy funkcji aproksymującej wynosi:",a[0])
```

```python
# 4.Rysowanie nałożonych na siebie wykresów

# Wykres punktowy dla wszystkich punktów
plt.scatter(dane[:,0],dane[:,1])

# Wykres funkcji aproksymującej
plt.plot(dane[:,0],a[0]*dane[:,0]+a[1],c='r')

# Legenda
plt.legend(['dane','funkcja aproksymująca'])
```
