```python
#  CZĘŚĆ 2
#  W folderze o nazwie czesc2 znajduje się 169 plików, z których każdy zawiera zbiór punktów. Dla każdego
#   z plików w folderze wykonaj następujące zadania.
#  1. Narysuj wykres liniowy, zawierający nałożone na siebie dane ze wszystkich plików z folderu czesc2.
#  2. Przeprowadź aproksymację liniową oraz wyznacz współczynnik kierunkowy funkcji aproksymującej dla wszystkich plików z 
#    folderu czesc2. Stwórz wektor, który będzie zawierał wyznaczone współczynniki kierunkowe dla wszystkich 
#    przeanalizowanych plików z tego folderu.
#  3. Oblicz średnią oraz odchylenie standardowe dla wektora, zawierającego współczynniki kierunkowe
#   funkcji.
#  4. Narysuj histogram dla wyznaczonych współczynników kierunkowych

# Importowanie potrzebnych bibliotek: numpy - wczytywanie danych,  matplotlib.pyplot - tworzenie wykresów, pandas - ładna ramka
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Wczytywanie danych z pliku

# Tworzymy listy zawierające współrzędne x i y dla każdego punktu
list_x = []
list_y = []

# Pętla do wczytywania plików
for i in range (1,169):
    
    # Dla każdego pliku wczytujemy współrzędne do "dane" po czym z danych pobieramy do list współrzędne x i y
    dane = np.loadtxt(f"dane/czesc2/punkty{i}.TXT",delimiter=',')
    
    list_x.append(dane[:,0])
    list_y.append(dane[:,1])
    
# W ten sposób list_x i list_y zawierają 169 list które mają wpółrzędne x i y, gdzie każda pomniejsza lista zawiera
#  punkty z jednego pliku

# Wykres liniowy
for i in range (0,168):
    
    plt.plot(list_x[i],list_y[i])
```

```python
# 2. Tworzenie wektora zawierającego współczynniki kierunkowe dla wszystkich plików

#Wektor na współczynniki
vector = np.zeros(168)

# W pętli najpierw przemiszczamy sie pomiędzy listami z plików po czym obliczamy wsp i dostajemy się do indeksu 0,
#  który zawiera współczynnik kierunkowy
for i in range (0,168):
    vector[i] = np.polyfit(list_x[i],list_y[i],1)[0]    
```

```python
# 3.Obliczanie średniej i odchylenia standardowego utworzonego wektora

# Tworzenie frame z biblioteki pandas z utworzonego wektora
pandzia = pd.DataFrame(vector, columns=['A'])

# Tworzenie zmiennej na statystyki
statystyki = pandzia.describe()

# Pobranie średniej i odchylenia ze statystyk
srednia = pandzia.describe().loc['mean','A']
odchylenie = pandzia.describe().loc['std','A']

# Wyświtlanie komunikatu o pobranych danych
print("Średnia wartości wektora wynosi:",srednia,", natomiast odchylenie standardowe wynosi:", odchylenie)
```

```python
# 4.Histogram 
plt.hist(vector)
plt.title("Histogram")
plt.show
```
